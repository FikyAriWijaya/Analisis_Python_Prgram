# **Menggabungkna Table**

```python
Data_Name = {
    'SSH' : [2,5,7,5,8],
    'Nama' : ['Fiky1', 'Fiky2', 'Fiky3', 'Fiky4', 'Fiky5']
}

Data_Umur = {
    'SSH' : [2,3,5,7,9],
    'Umur' : [20,21,22,23,24]
}
```

```python
data1 = pd.DataFrame(Data_Name)
data2 = pd.DataFrame(Data_Umur)
print(data1)
print('\n',data2)
```

```python
df = pd.merge(data1, data2, on = 'SSH', how = 'inner')
df.set_index('SSH', inplace = True)
df
```
