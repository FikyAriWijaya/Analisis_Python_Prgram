```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
x = np.arange(0, 100, 1)
y1 = np.sin(x)
y2 = x**2 - 2 * x
y3 = np.log(x)

ax1  = plt.subplot(2,2,1)
ax2 = plt.subplot(2,2,2)
ax3 = plt.subplot(2,2,3)
ax4 = plt.subplot(2,2,4)

ax1.plot(x,y1)
ax2.plot(x,y2)
ax3.plot(x,y1, 'r--')
ax4.plot(x,y2, 'go')
plt.tight_layout()
plt.show()
```

```python
aa = np.arange(0,30,0.2)
bb = np.sin(aa)
cc = np.cos(aa)

plt.figure(figsize = (10,4))
plt.title('Sine Fungsion')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(aa,bb, label = 'sin(x)')
plt.plot(aa,cc, label = 'cos(x)')
plt.legend(loc = 'upper right')

plt.show()
```

```python
python = (85, 70, 80, 83)
java = (90, 60, 80, 70)
networking = (70, 70, 80, 70)
machine_learning = (82, 80, 80, 85)

style.use('tableau-colorblind10')
people = ['Fky', 'John', 'Risky', 'Ronald']
index = np.arange(4)

plt.bar(index, python, width = 0.2, label = 'Python')
plt.bar(index + 0.2, java, width = 0.2, label = 'Java')
plt.bar(index + 0.4, networking, width = 0.2, label = 'Networking')
plt.bar(index + 0.6, machine_learning, width = 0.2, label = 'Machine_Learning')

plt.title('IT SKILL LEVEL')
plt.xticks(index + 0.3, people)
plt.legend(loc = 'upper right')
plt.ylim(0,150)

plt.show()
```

```python
nasionalisasi = ['Indonesia', 'Jepang', 'Korea', 'Turki']
numbersk = [60, 23, 21, 34]
explode  = [0,0,0.1,0]
plt.pie(numbersk, labels=nasionalisasi, autopct='%.2f%%', explode=explode)
plt.show()
```

```python
nasionalisasii = ['Indonesia', 'Jepang', 'Korea', 'Turki']
numberskk = [60, 23, 21, 34]
pairs = zip(numberskk,nasionalisasii)
pairs = sorted(list(pairs))
numberskk, nasionalisasii = zip(*pairs)

plt.pie(numberskk, labels=nasionalisasii, autopct='%.2f%%', counterclock = False, startangle = 90)
plt.title('Studen Nasionality')
plt.show()
```

```python
mu, sigma = 172, 8
x  = mu + sigma * np.random.random(1000)

plt.hist(x, 100, facecolor = 'Blue', density = True)
plt.xlabel('Height')
plt.ylabel('Probability')
plt.title('Height Distribution')
plt.text(172, 0.04, "Mu = 172, Sigma = 8")
plt.grid(True)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y, color = 'red', marker = 'x', s=100)
plt.show()
```

```python
ax = plt.axes(projection='3d')
z = np.linspace(0, 30, 100)
x = np.sin(z)
y = np.cos(z)

ax.plot3D(x,y,z, color = 'red')
plt.show()
```

```python
ax1 = plt.axes(projection='3d')

def fungsion_z(xa,ya):
  return np.sin(np.sqrt(xa**2 + ya**2))

xa = np.linspace(-5, 5, 100)
ya = np.linspace(-5, 5, 100)

xa,ya = np.meshgrid(xa,ya)
z = fungsion_z(xa,ya)

ax1.plot_surface(xa,ya,z, cmap = 'viridis')

plt.show()
```
