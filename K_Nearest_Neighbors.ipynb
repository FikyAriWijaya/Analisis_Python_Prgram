```python
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
data = load_breast_cancer()
print(data.feature_names)
print(data.target_names)
```

```python
x_train, x_test, y_train, y_test = train_test_split(np.array(data.data), np.array(data.target), test_size = 0.2)
```

```python
# Buat objek StandardScaler
scaler = StandardScaler()

# Fit dan transform x_train
x_train = scaler.fit_transform(x_train)

# Transform x_test
x_test = scaler.transform(x_test)
```

```python
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(x_train, y_train)
```

```python
print(clf.score(x_test, y_test))
```

```python
# Data baru
new_data_point = np.array([[10.0, 12.0, 65.0, 300.0, 0.1, 0.1, 0.05, 0.05, 0.1, 0.05,
                                   0.3, 0.4, 2.5, 40.0, 0.005, 0.02, 0.02, 0.01, 0.01, 0.005,
                                   12.0, 15.0, 70.0, 400.0, 0.12, 0.2, 0.15, 0.1, 0.2, 0.07]])
```

```python
# Standarisasi data titik baru
new_data_point = scaler.transform(new_data_point)
```
